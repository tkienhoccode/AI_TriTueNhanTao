```
#Bai 13:
def tron_mang_tong(a, b):
    kq = []
    i = j = 0
    n, m = len(a), len(b)

    while i < n and j < m:
        kq.append(a[i] + b[j])
        i += 1
        j += 1

    while i < n:
        kq.append(a[i])
        i += 1

    while j < m:
        kq.append(b[j])
        j += 1

    return kq

if __name__ == "__main__":
    a = list(map(int, input("Nhập mảng a: ").split()))
    b = list(map(int, input("Nhập mảng b: ").split()))
    print("Kết quả:", tron_mang_tong(a, b))
```

```
#Bai 14:
import random

def tao_mang(m, n, min_val=-10, max_val=10):
    return [[random.randint(min_val, max_val) for _ in range(n)] for _ in range(m)]

def xuat_dong(a, k):
    if 0 <= k < len(a):
        return a[k]
    return "Dòng không hợp lệ!"

def xuat_cot(a, k):
    n = len(a[0])
    if 0 <= k < n:
        return [a[i][k] for i in range(len(a))]
    return "Cột không hợp lệ!"

def dong_tong_lon_nhat(a):
    tong = [sum(row) for row in a]
    max_val = max(tong)
    return tong.index(max_val), max_val

def cot_tich_nho_nhat(a):
    m = len(a)
    n = len(a[0])
    tich = []

    for j in range(n):
        t = 1
        for i in range(m):
            t *= a[i][j]
        tich.append(t)

    min_val = min(tich)
    return tich.index(min_val), min_val

def dong_chan_cot_le(a):
    m = len(a)
    n = len(a[0])
    kq = []

    for i in range(m):
        for j in range(n):
            if i % 2 == 0 and j % 2 == 1:
                kq.append(a[i][j])

    return kq

def tbc_chan_dong_le(a):
    tong = 0
    dem = 0
    for i in range(len(a)):
        if i % 2 == 1:
            for val in a[i]:
                if val % 2 == 0:
                    tong += val
                    dem += 1

    return tong / dem if dem > 0 else 0

def tbc_bien(a):
    m = len(a)
    n = len(a[0])
    tong = 0
    dem = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1

    return tong / dem if dem > 0 else 0

def tbt_khong_bien(a):
    m = len(a)
    n = len(a[0])

    if m < 3 or n < 3:
        return 0

    tich = 1
    dem = 0

    for i in range(1, m-1):
        for j in range(1, n-1):
            tich *= a[i][j]
            dem += 1

    return tich ** (1/dem)

m = int(input("Nhập số dòng m: "))
n = int(input("Nhập số cột n: "))

a = tao_mang(m, n)

print("\nMảng vừa nhập:")
for row in a:
    print(row)

print("\nDòng k:", xuat_dong(a, int(input("Nhập dòng k: "))))
print("Cột k:", xuat_cot(a, int(input("Nhập cột k: "))))

dong, gt_dong = dong_tong_lon_nhat(a)
print(f"\nDòng có tổng lớn nhất: dòng {dong}, tổng = {gt_dong}")

cot, gt_cot = cot_tich_nho_nhat(a)
print(f"Cột có tích nhỏ nhất: cột {cot}, tích = {gt_cot}")

print("\nCác phần tử dòng chẵn – cột lẻ:", dong_chan_cot_le(a))
print("TBC các phần tử chẵn thuộc dòng lẻ:", tbc_chan_dong_le(a))
print("TBC các phần tử thuộc biên:", tbc_bien(a))
print("TBT các phần tử KHÔNG thuộc biên:", tbt_khong_bien(a))
```

```
#Bai 15:
class SinhVien:
    def __init__(self, ma, ten, nam_sinh, diem_tb):
        self.ma = ma
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb

def nhap_ds_sv(n):
    ds = []
    for i in range(n):
        print(f"\n--- Nhập sinh viên {i+1} ---")
        ma = input("Mã SV: ")
        ten = input("Tên: ")
        nam_sinh = int(input("Năm sinh: "))
        diem = float(input("Điểm TB: "))
        ds.append(SinhVien(ma, ten, nam_sinh, diem))
    return ds

def dem_len_lop(ds):
    return len([sv for sv in ds if sv.diem_tb >= 5])

def sinh_vien_20_tuoi(ds, nam_hien_tai):
    return [sv for sv in ds if nam_hien_tai - sv.nam_sinh >= 20]

def dem_he_dh(ds):
    return len([sv for sv in ds if sv.ma[2:4].upper() == "DH"])

def dem_ten_lan(ds):
    return len([sv for sv in ds if sv.ten.split()[-1].lower() == "lan"])

def dem_ho_phan(ds):
    return len([sv for sv in ds if sv.ten.split()[0].lower() == "phan"])


# --- Chương trình chính ---
n = int(input("Nhập số lượng sinh viên: "))
ds = nhap_ds_sv(n)

print("\nSố sinh viên lên lớp:", dem_len_lop(ds))

nam_hien_tai = int(input("\nNhập năm hiện tại: "))
print("\nDanh sách sinh viên đủ 20 tuổi:")
ds20 = sinh_vien_20_tuoi(ds, nam_hien_tai)

for sv in ds20:
    print(f" - Mã: {sv.ma}, Tên: {sv.ten}, Năm sinh: {sv.nam_sinh}, Điểm TB: {sv.diem_tb}")

print("\nSố sinh viên hệ DH:", dem_he_dh(ds))
print("Số sinh viên tên Lan:", dem_ten_lan(ds))
print("Số sinh viên họ Phan:", dem_ho_phan(ds))
```
